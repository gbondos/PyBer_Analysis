```python
# Import the dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the CSV file
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Plot the data frame 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Create a bar chart
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Create a pie, scatter chart by changing the plot.method
pyber_ride_df.plot.pie(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Alternatively you can use the kind method
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color="skyblue")
plt.ylabel("Avg. Fare ($USD)")
plt.show()
```
